## Load and prepare the dataset

### Subtask:
Load the dataset and prepare it for training a YOLOv8 model.


**Reasoning**:
Define the path to the dataset directory and load the dataset using the data.yaml file for YOLOv8 training.



```python
import os

# Define the path to the dataset directory
dataset_path = '/content/Accident-1'

# Verify the existence of data.yaml file
data_yaml_path = os.path.join(dataset_path, 'data.yaml')
if not os.path.exists(data_yaml_path):
    print(f"Error: data.yaml not found at {data_yaml_path}")
else:
    print(f"data.yaml found at {data_yaml_path}. Dataset is ready for YOLOv8 training.")
```

## Train the yolov8 model

### Subtask:
Train the model using the provided dataset.


**Reasoning**:
Import the YOLO class and train the model using the specified dataset path and number of epochs.



```python
from ultralytics import YOLO

# Load a pre-trained YOLOv8 nano model
model = YOLO('yolov8n.pt')

# Train the model
model.train(data=dataset_path + '/data.yaml', epochs=50)
```

## Validate the model

### Subtask:
Evaluate the performance of the trained model.

**Reasoning**:
Evaluate the trained YOLOv8 model on the validation dataset using the `validate()` method.

```python
# Validate the trained model
metrics = model.val()

# Print the validation metrics
print(metrics)
```

Implement a function to send an SMS notification using Twilio. This function will be called when a crash is detected.

```python
from twilio.rest import Client
import os

account_sid = os.environ.get("TWILIO_ACCOUNT_SID")
auth_token = os.environ.get("TWILIO_AUTH_TOKEN")

if not account_sid or not auth_token:
    print("Twilio Account SID and Auth Token not found. Please add them to Colab Secrets.")
else:
    client = Client(account_sid, auth_token)

    def send_accident_notification(to_phone_number, body):
        """Sends an SMS notification using Twilio."""
        try:
            message = client.messages.create(
                to=to_phone_number,
                from_="you phone number",
                body=body)
            print(f"Notification sent successfully: {message.sid}")
        except Exception as e:
            print(f"Error sending notification: {e}")

```
